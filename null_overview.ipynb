To better view null values, drop the following columns:
- country: redundant information, i.e. same info as country_long
- name: will be using gppd_indr as identifier for this visualization
- latitude: unecessary info for this analysis
- longitude: unecessary info for this analysis
- url: unecessary info for this analysis
- geolocation_source: unecessary info for this analysis

```python
remove_cols = ['country', 'name', 'latitude', 'longitude', 'url', 'geolocation_source', 'wepp_id', 'generation_data_source']
gpp_removed = gpp_data.drop(columns= remove_cols)
gpp_removed.head()
```

```python
sns.heatmap(gpp_removed.isnull(), cmap= 'viridis')
plt.show()
```
